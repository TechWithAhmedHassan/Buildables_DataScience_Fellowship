```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. NumPy Array Creation
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("Array 1:", arr1)
print("Array 2:\n", arr2)
print("Addition:", arr1 + 10)
print("Multiplication:", arr1 * 2)
print("Slicing:", arr2[:, 1])

# 2. Array Statistics
print("Mean:", np.mean(arr1))
print("Median:", np.median(arr1))
print("Std:", np.std(arr1))
print("Argmax:", np.argmax(arr1))

# 3. Pandas Series & DataFrame
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print("\nSeries:\n", s)
data = {"Name": ["Ali", "Sara", "Omar"], "Age": [25, 28, 30]}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)

# 4. Data Importing (demo CSV creation for practice)
df.to_csv("data.csv", index=False)
df_csv = pd.read_csv("data.csv")
print("\nCSV Head:\n", df_csv.head())
print("CSV Tail:\n", df_csv.tail())

# 5. Data Cleaning
df_nan = pd.DataFrame({"A": [1, np.nan, 3], "B": [4, 5, np.nan]})
print("\nOriginal with NaN:\n", df_nan)
print("Fillna:\n", df_nan.fillna(0))
print("Dropna:\n", df_nan.dropna())

# 6. Indexing & Selection
print("Loc:", df.loc[0, "Name"])
print("Iloc:", df.iloc[0, 1])
print("Column:", df.loc[:, "Age"])

# 7. Sorting & Filtering
print("\nFiltered (Age>25):\n", df[df["Age"] > 25])
print("Sorted by Age:\n", df.sort_values("Age", ascending=False))

# 8. GroupBy & Aggregations
df_sales = pd.DataFrame({
    "Category": ["A", "B", "A", "B"],
    "Sales": [100, 200, 150, 250]
})
print("\nGroupBy Avg Sales:\n", df_sales.groupby("Category")["Sales"].mean())

# 9. Merging & Joining
df1 = pd.DataFrame({"ID": [1, 2], "Name": ["Ali", "Sara"]})
df2 = pd.DataFrame({"ID": [1, 2], "Score": [85, 90]})
merged = pd.merge(df1, df2, on="ID")
print("\nMerged DF:\n", merged)

# 10. Descriptive Statistics
print("\nDescribe:\n", df_sales.describe())



# 11. Line Plot Basics
values = [1, 3, 2, 5, 7]
plt.plot(values)
plt.title("Line Plot Example")
plt.show()

# 12. Bar Chart Frequencies
categories = ["A", "B", "C"]
counts = [5, 7, 3]
plt.bar(categories, counts)
plt.title("Bar Chart Example")
plt.show()

# 13. Scatter Plot Relationships
x = [5, 7, 8, 7, 6]
y = [99, 86, 87, 88, 100]
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.show()

# 14. Seaborn Pairplot
iris = sns.load_dataset("iris")
sns.pairplot(iris, hue="species")
plt.show()

# 15. Category Box Plot
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Box Plot Example")
plt.show()

# 16. Correlation Heatmap
corr = iris.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 17. Subplot Grid
fig, ax = plt.subplots(2, 2, figsize=(8, 6))
ax[0,0].plot(values)
ax[0,1].bar(categories, counts)
ax[1,0].scatter(x, y)
ax[1,1].hist(iris["sepal_length"])
plt.tight_layout()
plt.show()

# 18. Seaborn Styling
sns.set_style("whitegrid")
sns.barplot(x=categories, y=counts, palette="muted")
plt.title("Custom Bar Plot")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()

# 19. Violin Plot Distribution
sns.violinplot(x="species", y="sepal_length", data=iris)
plt.title("Violin Plot")
plt.show()

# 20. Hybrid Visualization
sns.histplot(iris["sepal_length"], kde=True, bins=20)
plt.title("Histogram + KDE")
plt.show()
```
